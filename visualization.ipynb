```python
import cv2
import os
import json
from collections import defaultdict
import random
```

```python
img_path = "/home/tanluuuuuuu/Desktop/luunvt/competitions/VIPriors Object Detection challenge/scr/datasets/DelftBikes/images"
label_path = "/home/tanluuuuuuu/Desktop/luunvt/competitions/VIPriors Object Detection challenge/scr/datasets/DelftBikes/train_anno_COCO.json"
visualization_path = "/home/tanluuuuuuu/Desktop/luunvt/competitions/VIPriors Object Detection challenge/scr/datasets/visualization"
```

```python
f = open(label_path)
labels = json.load(f)
```

```python
id_to_img_name = defaultdict()
for image in labels['images']:
    id_to_img_name[image['id']] = image['file_name']
print(id_to_img_name)
```

```python
dict_images = defaultdict(list)


for annotation in labels['annotations']:
    image_id = annotation['image_id']
    bbox = annotation['bbox']
    category_id = annotation['category_id']
    image_name = id_to_img_name[image_id]
    dict_images[image_name].append(bbox)
print(dict_images['48678.jpg'])
```

```python
def yolobbox2bbox(x,y,w,h):
    x1, y1 = x-w/2, y-h/2
    x2, y2 = x+w/2, y+h/2
    return int(x1), int(y1), int(x2), int(y2)
```

```python
names = [  
  "back_hand_break",
  "back_handle",
  "back_light",
  "back_mudguard",
  "back_pedal",
  "back_reflector",
  "back_wheel",
  "bell",
  "chain",
  "dress_guard",
  "dynamo",
  "front_handbreak",
  "front_handle",
  "front_light",
  "front_mudguard",
  "front_pedal",
  "front_wheel",
  "gear_case",
  "kickstand",
  "lock",
  "saddle",
  "steer"
    ]
colors = [[random.randint(0, 255) for _ in range(3)] for _ in names]
colors
```

```python
for image_name in dict_images:
    full_image_path = os.path.join(img_path, image_name)
    full_output_path = os.path.join(visualization_path, image_name)

    bboxes = dict_images[image_name]
    image = cv2.imread(full_image_path)
    for id, box in enumerate(bboxes):
        x, y, w, h = box
        x1, y1, x2, y2 = yolobbox2bbox(x, y, w, h)
        image = cv2.rectangle(image, (x1, y1), (x2, y2), colors[id], 2)

        tl = round(0.002 * (image.shape[0] + image.shape[1]) / 2) + 1  # line/font thickness
        tf = max(tl - 1, 1)  # font thickness
        c1, c2 = (int(x1), int(y1)), (int(x2), int(y2))
        t_size = cv2.getTextSize(names[id], 0, fontScale=tl / 3, thickness=tf)[0]
        c2 = x1 + t_size[0], y1 - t_size[1] - 3
        image = cv2.rectangle(image, c1, c2, colors[id], -1, cv2.LINE_AA)  # filled
        image = cv2.putText(image, names[id], (x1, y1 - 2), 0, tl / 3, [225, 255, 255], thickness=tf, lineType=cv2.LINE_AA)
    cv2.imwrite(full_output_path, image)
```
